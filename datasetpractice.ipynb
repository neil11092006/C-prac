```python
import pandas as pd
import numpy as np

df = pd.read_csv('clinvar_conflicting.csv', low_memory=False)
df.head()
```

```python
#Unify the Chromosome Column to avoid mixed text/number errors
df['CHROM'] = df['CHROM'].astype(str)
df.head()
```

```python
#Impute missing population frequencies with 0.0 for rare varients
af_columns = [col for col in df.columns if col.startswith('AF_')]
df[af_columns] = df[af_columns].fillna(0)
df.head()
```

```python
#Drop uninformative columns missimg more than 99% of their data
motif_columns = [col for col in df.columns if col.startswith('MOTIF_')]
df = df.drop(columns=motif_columns)
df.head()
```

```python
df = df.drop(columns=['HIGH_INF_POS'])
df.head()
```

```python
#upgrade2 2.0 = Identify all columns with more than 90% missing values
missing_threshold = 0.90
empty_columns = df.columns[df.isna().mean() > missing_threshold]
```

```python
#drop them all with one single command
df = df.drop(columns=empty_columns)

print(f"Dropped columns with more than 90% missing values: {empty_columns}")
print(f"Final clean dataset dimensions: {df.shape}")
```
